```python
class hotelcharges:

    room_no_count = 0

    def __init__(
        self,
        rt="",
        s=0,
        p=0,
        r=0,
        t=0,
        a=1800,
        name="",
        address="",
        cindate="",
        coutdate="",
        rno=0,
    ):

        print("*****WELCOME TO TAJ HOTEL*****")

        self.rt = rt

        self.r = r

        self.t = t

        self.p = p

        self.s = s
        self.a = a
        self.name = name
        self.address = address
        self.cindate = cindate
        self.coutdate = coutdate
        self.rno = hotelcharges.room_no_count + 1
        hotelcharges.room_no_count = hotelcharges.room_no_count + 1

    def inputdata(self):
        self.name = input("Enter your name:")
        self.address = input("Enter your address:")
        self.cindate = input("Enter your check in date:")
        self.coutdate = input("Enter your checkout date:")
        print("Your room no.:", self.rno, "\n")

    def roomcharge(self):

        print("We have the following rooms for you:-")

        print("1. Deluxe Room---->₹ 6000 per night\-")

        print("2. Family Room---->₹ 5000 per night\-")

        print("3. Standard Room---->₹ 4000 per night\-")

        print("4. Regular Room---->₹ 3000 per night\-")

        x = int(input("Enter Your Choice Please->"))

        n = int(input("For How Many Nights Did Customer Stayed:"))

        if x == 1:

            print("You have opted for Deluxe Room")

            self.s = 6000 * n

        elif x == 2:

            print("You have opted for Family Room")

            self.s = 5000 * n

        elif x == 3:

            print("You have opted for Standard Room")

            self.s = 4000 * n

        elif x == 4:
            print("You have opted for Regular Room")

            self.s = 3000 * n

        else:

            print("Please choose a room")

        print("Your room charge is = ₹", self.s, "\n")

    def restaurantbill(self):

        print("*****RESTAURANT MENU*****")

        print(
            "1.Water----->₹ 20",
            "2.Tea----->₹ 20",
            "3.Coffee----->₹ 30",
            "4.Breakfast--->₹ 250",
            "5.Lunch---->₹ 400",
            "6.Dinner--->₹ 550",
            "7.Exit",
            )
        while 1:

            c = int(input("Enter your choice:"))

            if c == 1:
                d = int(input("Enter the quantity:"))
                self.r = self.r + 20 * d

            elif c == 2:
                d = int(input("Enter the quantity:"))
                self.r = self.r + 20 * d

            elif c == 3:
                d = int(input("Enter the quantity:"))
                self.r = self.r + 30 * d

            elif c == 4:
                d = int(input("Enter the quantity:"))
                self.r = self.r + 250 * d

            elif c == 5:
                d = int(input("Enter the quantity:"))
                self.r = self.r + 400 * d

            elif c == 6:
                d = int(input("Enter the quantity:"))
                self.r = self.r + 550 * d
        
            elif c == 7:
                break

            else:
                print("Invalid option")

        print("Total food bill= ₹", self.r, "\n")

    def laundrybill(self):
        print("******LAUNDRY MENU*******")

        print(
            "1.Shorts----->₹ 15",
            "2.Trousers----->₹ 25",
            "3.Shirt--->₹ 20",
            "4.Jeans---->₹ 30",
            "5.Sarees--->₹ 30",
            "6.Exit",
        )

        while 1:

            e = int(input("Enter your choice:"))

            if e == 1:
                f = int(input("Enter the quantity:"))
                self.t = self.t + 15 * f

            elif e == 2:
                f = int(input("Enter the quantity:"))
                self.t = self.t + 25 * f

            elif e == 3:
                f = int(input("Enter the quantity:"))
                self.t = self.t + 20 * f

            elif e == 4:
                f = int(input("Enter the quantity:"))
                self.t = self.t + 30 * f

            elif e == 5:
                f = int(input("Enter the quantity:"))
                self.t = self.t + 30 * f
            elif e == 6:
                break
            else:

                print("Invalid option")

        print("Total Laundary Bill= ₹", self.t, "\n")

    def barbill(self):
        print("******BAR MENU*******")

        print(
            "1.Wine----->₹ 650",
            "2.Bear----->₹ 400",
            "3.Whisky--->₹ 500",
            "4.Cocktails---->₹ 600",
            "5.Mocktails--->₹ 300",
            "6.Exit",
        )

        while 1:

            g = int(input("Enter your choice:"))

            if g == 1:
                h = int(input("Enter the quantity:"))
                self.p = self.p + 650 * h

            elif g == 2:
                h = int(input("Enter the quantity:"))
                self.p = self.p + 400 * h

            elif g == 3:
                h = int(input("Enter the quantity:"))
                self.p = self.p + 500 * h

            elif g == 4:
                h = int(input("Enter the quantity:"))
                self.p = self.p + 600 * h

            elif g == 5:
                h = int(input("Enter the quantity:"))
                self.p = self.p + 300 * h
            elif g == 6:
                break

            else:

                print("Invalid option")

        print("Total Bar Bill= ₹", self.p, "\n")

    def display(self):
        print("******HOTEL BILL******")
        print("Customer details:")
        print("Customer name:", self.name)
        print("Customer address:", self.address)
        print("Check in date:", self.cindate)
        print("Check out date", self.coutdate)
        print("Room no.", self.rno)
        print("Your Room charge is:", self.s)
        print("Your Food bill is:", self.r)
        print("Your Laundary bill is:", self.t)
        print("Your Bar bill is:", self.p)

        self.rt = self.s + self.t + self.p + self.r

        print("Your sub total bill is:", self.rt)
        print("Additional Service Charges is", self.a)
        print("Your grandtotal bill is:", self.rt + self.a, "\n")


def main():

    rooms = []
    a = None
    while 1:
        print("1.Enter Customer Data")

        print("2.Calculate room charge")

        print("3.Calculate restaurant bill")

        print("4.Calculate laundry bill")

        print("5.Calculate bar bill")

        print("6.Show total bill")

        print("7.Rooms booked")

        print("8.EXIT")

        b = int(input("Enter your choice:"))
        if b == 1:
            a = hotelcharges()
            rooms.append({"room_no": a.rno})
            a.inputdata()
        if b == 2:
            try:
                a.roomcharge()
            except Exception as e:
                print("Please enter Customer data first")
        if b == 3:
            try:
                a.restaurantbill()
            except Exception as e:
                print("Please enter Customer data first")
        if b == 4:
            try:
                a.laundrybill()
            except Exception as e:
                print("Please enter Customer data first")
        if b == 5:
            try:
                a.barbill()
            except Exception as e:
                print("Please enter Customer data first")
        if b == 6:
            try:
                a.display()
            except Exception as e:
                print("Please enter Customer data first")
        if b == 7:
            print(rooms) if len(rooms) > 0 else print("All rooms available")
        if b == 8:

            quit()


main()
```
